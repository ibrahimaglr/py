```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
df=pd.read_csv("train.csv")
df.tarih=df.tarih.apply(pd.to_datetime)
df['ts'] = df[['tarih']].apply(lambda x: x[0].timestamp(), axis=1)
```

```python
le=LabelEncoder()
df["ürün"]=le.fit_transform(df["ürün"])
df["ürün üretim yeri"]=le.fit_transform(df["ürün üretim yeri"])
df["market"]=le.fit_transform(df["market"])
df["şehir"]=le.fit_transform(df["şehir"])
```

```python
x=df.loc[:,["ürün","market","şehir","ürün üretim yeri"]]
ohe= preprocessing.OneHotEncoder()
x3=ohe.fit_transform(x).toarray()
```

```python
x=pd.DataFrame(x3)
x2=df.loc[:,["ürün fiyatı","ts"]]
df_2=pd.concat([x,x2],axis=1)
```

```python
r3=[]
x=df_2.drop(columns=['ürün fiyatı'])
y=df_2["ürün fiyatı"]
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=42)

rf = RandomForestRegressor(n_estimators=1000)
rf.fit(x,y)
```

```python
df_test=pd.read_csv("testFeatures.csv")
df_test.tarih=df_test.tarih.apply(pd.to_datetime)
df_test['ts'] = df_test[['tarih']].apply(lambda x: x[0].timestamp(), axis=1)
df_test["ürün"]=le.fit_transform(df_test["ürün"])
df_test["ürün üretim yeri"]=le.fit_transform(df_test["ürün üretim yeri"])
df_test["market"]=le.fit_transform(df_test["market"])
df_test["şehir"]=le.fit_transform(df_test["şehir"])
x2=df_test.loc[:,["ürün","market","şehir","ürün üretim yeri"]]
ohe= preprocessing.OneHotEncoder()
x3=ohe.fit_transform(x2).toarray()
x=pd.DataFrame(x3)
x5=df_test.loc[:,["ts"]]
df_2=pd.concat([x,x5],axis=1)
```
